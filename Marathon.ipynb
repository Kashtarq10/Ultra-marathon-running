```python
#clean up data
```

```python
#only USA races 50km or 50miles in 2020
```

```python
#show 50 km or 50 Mile


df[df['Event distance/length'] == '50mi']
```

```python
# combine 50km and 50mi

df[df['Event distance/length'].isin(['50mi','50k'])]
```

```python
# now on; for year 2020 and in USA

df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020) ]
```

```python
df.head(100)
```

```python
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2= df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#Gte age
df2['athelet_age']= 2020 - df2['Athlete year of birth']
```

```python
# remove 'h' from performance to clean it
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

```python
#drop club, country, DOB, category

df2 = df2.drop(['Athlete club','Athlete country' ,'Athlete year of birth','Athlete age category'], axis =1)
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
df2.isna().sum()
```

```python
df2= df2.dropna()
```

```python
df2.shape
```

```python
df2= [df2.duplicate()== True]
```

```python
df2.shape
```

```python
#reset index

df2.reset_index(drop= True)
```

```python
#fix_type
df2.dtypes
```

```python
df2['athelet_age']= df2['athelet_age'].astype(int)
```

```python
df2['Athlete average speed']= df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#rename columns
            
```

```python
df2= df2.rename(columns ={'Year of event': 'Event_year' ,          
                            'Event dates' : 'Event_dates',                    
                            'Event name'  :  'Event_name',                 
                            'Event distance/length' : 'Distance',         
                            'Event number of finishers'  :'Event_finishers' ,      
                            'Athlete performance'  : 'Athlete_performance' ,          
                            'Athlete gender'   :  'Athlete_gender',              
                            'Athlete average speed'   :'Athlete_average_speed' ,       
                            'Athlete ID'  :'ID' ,                     
                            'athelet_age'   :'Age'})                
```

```python
df2.head(100)
```

```python
# findd 2 particular races in 2020
df2[df2['Event_name'] == 'West Seattle Beach Run - Winter Edition ']
```

```python
#Question 1 Difference in speed for 50km and 50 mi for male and female.

df2.groupby(['Distance' , 'Athlete_gender'])['Athlete_average_speed'].mean()
```

```python
# Shows Distance in (km and miles) count

sns.histplot(df2['Distance'])
```

```python
# Shows Gender VS Distance (km and miles)

sns.histplot(df2, x= 'Distance', hue = 'Athlete_gender')
```

```python
# Shows Athlete_average_speed

sns.histplot(df2[df2['Distance'] == '50mi']['Athlete_average_speed'])
```

```python
# Shows Gender comapnraion in 50km and 50 mi races
sns.violinplot(data = df2, x = 'Distance', y = 'Athlete_average_speed', hue = 'Athlete_gender', split = True ,inner ='quart', linewidth =1)
```

```python
#Shows Different age group Vs Speed

sns.lmplot( data= df2 , x ='Age' , y = 'Athlete_average_speed', hue = 'Athlete_gender')
```
